```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# from matplotlib_venn import venn2
```

```python
df1 = pd.read_csv('../datasets/1k_data.csv', index_col = 0)
df2 = pd.read_csv('../datasets/2k_data.csv', index_col = 0)
df3 = pd.read_csv('../datasets/3k_data.csv', index_col = 0)
df4 = pd.read_csv('../datasets/4k_data.csv', index_col = 0)
df5 = pd.read_csv('../datasets/5k_data.csv', index_col = 0)
```

```python
game_df = pd.concat([df1,df2,df3,df4,df5]).drop_duplicates().reset_index(drop=True)
```

```python
# NEED TO AUTOMATE

# subset list of team compositions + frequency for players with high placement (1)
placement_1_df = game_df[game_df['placement'] == 1]
placement_1_df = placement_1_df.sum()
placement_1_df = pd.DataFrame(placement_1_df)
placement_1_list = placement_1_df.iloc[14:234]

# remove range = 0 REMEMBER
#for num in range(8)

# subset list of team compositions + frequency for players with average placement (2)
placement_2_df = game_df[game_df['placement'] == 2]
placement_2_df = placement_2_df.sum()
placement_2_df = pd.DataFrame(placement_2_df)
placement_2_list = placement_2_df.iloc[14:234]

placement_3_df = game_df[game_df['placement'] == 3]
placement_3_df = placement_3_df.sum()
placement_3_df = pd.DataFrame(placement_3_df)
placement_3_list = placement_3_df.iloc[14:234]

placement_4_df = game_df[game_df['placement'] == 4]
placement_4_df = placement_4_df.sum()
placement_4_df = pd.DataFrame(placement_4_df)
placement_4_list = placement_4_df.iloc[14:234]


# ???????????
placement_5_df = game_df[game_df['placement'] == 5]
placement_5_df = placement_5_df.sum()
placement_5_df = pd.DataFrame(placement_5_df)
placement_5_list = placement_5_df.iloc[14:234]

placement_6_df = game_df[game_df['placement'] == 6]
placement_6_df = placement_6_df.sum()
placement_6_df = pd.DataFrame(placement_6_df)
placement_6_list = placement_6_df.iloc[14:234]

placement_7_df = game_df[game_df['placement'] == 7]
placement_7_df = placement_7_df.sum()
placement_7_df = pd.DataFrame(placement_7_df)
placement_7_list = placement_7_df.iloc[14:234]

placement_8_df = game_df[game_df['placement'] == 8]
placement_8_df = placement_8_df.sum()
placement_8_df = pd.DataFrame(placement_8_df)
placement_8_list = placement_8_df.iloc[14:234]
```

```python
# change the index and sort values to have individual lists of most used combinations

placement_1_list = placement_1_list.reset_index()
placement_1_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_1_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_2_list = placement_2_list.reset_index()
placement_2_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_2_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_3_list = placement_3_list.reset_index()
placement_3_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_3_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_4_list = placement_4_list.reset_index()
placement_4_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_4_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_5_list = placement_5_list.reset_index()
placement_5_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_5_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_6_list = placement_6_list.reset_index()
placement_6_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_6_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_7_list = placement_7_list.reset_index()
placement_7_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_7_list.sort_values(by = 'count', ascending=False, inplace = True)

placement_8_list = placement_8_list.reset_index()
placement_8_list.rename( columns={'index' : 'combination', 0 :'count'}, inplace=True )
placement_8_list.sort_values(by = 'count', ascending=False, inplace = True)
```
